```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
df = pd.merge(study_results, mouse_metadata, how = 'left', on=['Mouse ID', 'Mouse ID'])
df = df.rename(columns={'Tumor Volume (mm3)': 'Volume'})

# Display the data table for preview
df
```

```python
# Checking the number of mice.
df['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate = df[df.duplicated(['Mouse ID', 'Timepoint'])]

duplicate
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_df = df.set_index('Mouse ID')
duplicate_df.loc['g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = duplicate_df.drop(['g989'])
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
drop_clean_df = clean_df.reset_index()
drop_clean_df['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drug_df = drop_clean_df.groupby(['Drug Regimen'])
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mean = drug_df['Volume'].mean()
med = drug_df['Volume'].median()
var = drug_df['Volume'].var()
std = drug_df['Volume'].std()
sem = drug_df['Volume'].sem()

tumor_df = pd.DataFrame({
    'Mean': mean,
    'Median': med,
    'Variance': var,
    'Standard Deviation': std,
    'SEM': sem
})

tumor_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
tumor_df2 = drop_clean_df.groupby(['Drug Regimen'])['Volume'].agg(['mean', 'median', 'var', 'std', 'sem'])

tumor_df2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_tumor = drop_clean_df.groupby(['Mouse ID'])['Timepoint'].max()
final_tumor_df = pd.DataFrame(final_tumor)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor = pd.merge(drop_clean_df, final_tumor_df, on=('Mouse ID', 'Timepoint'), how='right')
final_tumor

```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
cap = []
ram = []
inf = []
cef = []


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes
for index, row in final_tumor.iterrows():
    if row['Drug Regimen'] == 'Capomulin':
        ram.append(row['Volume'])
    if row['Drug Regimen'] == 'Ramicane':
        cap.append(row['Volume'])
    if row['Drug Regimen'] == 'Infubinol':
        inf.append(row['Volume'])
    if row['Drug Regimen'] == 'Ceftamin':
        cef.append(row['Volume'])
    
    # add subset 
treatment_df = pd.DataFrame({treatments[0]: cap,
                            treatments[1]: ram,
                            treatments[2]: inf,
                            treatments[3]: cef})

    
    # Determine outliers using upper and lower bounds
cap_quart = treatment_df[treatments[0]].quantile([.25, .5, .75])
cap_low = cap_quart[.25]
cap_high = cap_quart[.75]
cap_iqr = cap_high-cap_low
cap_lowerb = cap_low - (1.5*cap_iqr)
cap_upperb = cap_high + (1.5*cap_iqr)
cap_outliers = [data for data in treatment_df[treatments[0]] if data < cap_lowerb or data > cap_upperb]
# sorry for the hard code, was taking too long trying to simplfy it


ram_quart = treatment_df[treatments[1]].quantile([.25, .5, .75])
ram_low = ram_quart[.25]
ram_high = ram_quart[.75]
ram_iqr = ram_high-cap_low
ram_lowerb = ram_low - (1.5*ram_iqr)
ram_upperb = ram_high + (1.5*ram_iqr)
ram_outliers = [data for data in treatment_df[treatments[1]] if data < ram_lowerb or data > ram_upperb]

inf_quart = treatment_df[treatments[2]].quantile([.25, .5, .75])
inf_low = inf_quart[.25]
inf_high = inf_quart[.75]
inf_iqr = inf_high-cap_low
inf_lowerb = inf_low - (1.5*inf_iqr)
inf_upperb = inf_high + (1.5*inf_iqr)
inf_outliers = [data for data in treatment_df[treatments[2]] if data < inf_lowerb or data > inf_upperb]

cef_quart = treatment_df[treatments[3]].quantile([.25, .5, .75])
cef_low = cef_quart[.25]
cef_high = cef_quart[.75]
cef_iqr = cef_high-cap_low
cef_lowerb = cef_low - (1.5*cef_iqr)
cef_upperb = cef_high + (1.5*cef_iqr)
cef_outliers = [data for data in treatment_df[treatments[3]] if data < cef_lowerb or data > cef_upperb]

outliers = []
outliers.append(cap_outliers)
outliers.append(ram_outliers)
outliers.append(inf_outliers)
outliers.append(cef_outliers)
outliers
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
drug_list = [treatment_df[treatments[0]], 
             treatment_df[treatments[1]], 
             treatment_df[treatments[2]], 
             treatment_df[treatments[3]]]

fig, ax = plt.subplots(figsize = (9,7))
ax.set_title('Final Measured Tumor Volume by Drug Regimen')
ax.set_xlabel('Drug Regimen')
ax.set_ylabel('Tumor Volume (mm3)')

ax.boxplot(drug_list, 0, 'gD')
plt.xticks([1,2,3,4], treatments)
plt.savefig('Boxplot')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
cap_df = drop_clean_df.loc[drop_clean_df["Drug Regimen"] == "Capomulin",:]

cap_avg = cap_df.groupby(['Mouse ID']).mean()

plt.scatter(cap_avg['Weight (g)'],cap_avg['Volume'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

plt.savefig('Scatterplot')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
corr = round(st.pearsonr(cap_avg['Weight (g)'],cap_avg['Volume'])[0],2)
print(f'The correlation between mouse weight and average tumor volume is {corr}')


# From regress variable
slope = 0.9544396890241049
intercept = 21.552160532685008

line = slope * (cap_avg['Weight (g)']) + intercept

plt.scatter(cap_avg['Weight (g)'],cap_avg['Volume'])

plt.plot(cap_avg['Weight (g)'], line, 'r', label = 'y={:.2f}x+{:.2f}'.format(slope, intercept))
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

plt.legend()
plt.savefig('Linear Regression')
plt.show()
```
